```python
#Step 2:  Write the Code to Pass the Test
#Modify the test to correctly capture the output of the plot and ensure it passes.



class TestPlotStockData(unittest.TestCase):
    def test_plot_stock_data_contains_lines(self):
        df = pd.read_csv('data/AMZN_2024-05-19.csv')
        fig, ax = plt.subplots()
        plot_stock_data(df)
        self.assertTrue(len(ax.lines) > 0, "The plot should contain lines.")

    def test_plot_stock_data_labels_and_title(self):
        df = pd.read_csv('data/AMZN_2024-05-19.csv')
        fig, ax = plt.subplots()
        plot_stock_data(df)
        self.assertEqual(ax.get_xlabel(), 'Date', "The x-axis label should be 'Date'.")
        self.assertEqual(ax.get_ylabel(), 'Value', "The y-axis label should be 'Value'.")
        self.assertEqual(ax.get_title(), 'Stock Data', "The title should be 'Stock Data'.")

    def test_plot_stock_data_columns_present(self):
        df = pd.read_csv('data/AMZN_2024-05-19.csv')
        fig, ax = plt.subplots()
        plot_stock_data(df)
        line_labels = [line.get_label() for line in ax.get_lines()]
        expected_labels = ['Open', 'High', 'Low', 'Close', 'Adjusted Close', 'Volume']
        self.assertTrue(all(label in line_labels for label in expected_labels), "All columns should be present in the plot.")

# Run the tests
unittest.main(argv=[''], verbosity=2, exit=False)
```
